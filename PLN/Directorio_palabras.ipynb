El objetivo es realizar un conteo de las palabras de un texto, el archivo de entrada puede ser de cualquier formato, pero debe de entregar un histograma del mismo.
Realizar una nube de palabras a partir del corpus obtenido de su base de datos de proyecto o el corpus de cualquier libro.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

Descarga los recursos necesarios de NLTK

```python
nltk.download('punkt')
```

Lectura del archivo de texto

```python
with open('txtSample.txt', 'r', encoding='utf-8') as file:
    texto = file.read()
```

Filtro de palabras, para evitar signos como puntos o comas, y conteo de las palabras

```python
words = word_tokenize(texto)
filtered_words = [i.lower() for i in words if len(i) > 3]
freq = FreqDist(filtered_words)
```

Obtenci칩n del histograma con las 20 palabras m치s comunes

```python
most_common_words = freq.most_common(20)
words, freqs = zip(*most_common_words)

plt.figure(figsize=(8, 4))
plt.bar(words, freqs)
plt.title('Histograma de Frecuencia de Palabras')
plt.xlabel('Palabras')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(words))
plt.figure(figsize=(8, 4))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Nube de Palabras')
plt.show()
```
